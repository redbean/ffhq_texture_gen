```python

import os, sys
import numpy as np
from tqdm import tqdm
```

```python
target_dir = os.path.join(os.getcwd(), 'data_prep_4chan')
```

```python
file = os.listdir(target_dir)
out_dir = os.path.join(os.getcwd(), "data_prep_9wh")
if not os.path.exists(out_dir):
    os.makedirs(out_dir)
```

```python

for i, file in enumerate(tqdm(file)):
    path = os.path.join(target_dir, file)
    data = np.load(path)
    output_data = data.reshape(12, data.shape[3], data.shape[3])    
    np.save(os.path.join(out_dir, f'data_{i:06}.npy'), output_data)
```

```python
import matplotlib.pyplot as plt
output_array = np.load("test.npy")

# 첫 세 개의 채널을 합쳐 하나의 RGB 이미지로 변환
rgb_image = np.stack((output_array[0], output_array[1], output_array[2]), axis=-1)

# 이미지 시각화
plt.imshow(rgb_image)
plt.title('Combined RGB Image from First Three Channels')
plt.show()
```
